```python
from collections import deque

graph = {
    1: [2, 3, 4],
    2: [5, 6],
    3: [],
    4: [7, 8],
    5: [9, 10],
    6: [],
    7: [11, 12],
    8: [],
    9: [],
    10: [],
    11: [],
    12: []
}

destination = 12
current = 1
path = [1]
all_nodes = set(graph.keys())

def has_path_to_L(start):
    queue = deque([start])
    visited = set([start])
    while queue:
        node = queue.popleft()
        if node == destination:
            return True
        for child in graph[node]:
            if child not in visited:
                visited.add(child)
                queue.append(child)
    return False

def print_state(current, path, selected):
    non_visited = sorted(all_nodes - set(path))
    print(f"\nCurrent Node: {current}")
    print("Vertices (children):", ", ".join(map(str, graph[current])))
    print("Edges:", ", ".join(f"({current},{v})" for v in graph[current]))
    print("Visited Nodes:", ", ".join(map(str, path)))
    print("Non-Visited Nodes:", ", ".join(map(str, non_visited)))
    print(f"SELECTED Node: {selected}")
    print("PATH:", "".join(map(str, path)))

for step in range(3):
    for child in graph[current]:
        if step < 2 and has_path_to_L(child):
            selected = child
            break
        elif step == 2 and child == destination:
            selected = child
            break
    path.append(selected)
    print_state(current, path, selected)
    current = selected

print("\nSHORTEST PATH FROM 1 TO 12:")
print(" → ".join(map(str, path)))
```
